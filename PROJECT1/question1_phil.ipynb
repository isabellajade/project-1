What are the best locations by review category (accuracy, cleanliness, checkin, communication, location, and value)?

```python
# import dependencies 
import pandas as pd
import numpy as np
import hvplot.pandas
import matplotlib.pyplot as plt
```

```python
# read combined csv
combined_df = pd.read_csv('resources/combined_data.csv')

# reduce data to only locations with hot tubs
hot_tubs_df = combined_df[combined_df['has_hot_tub'] == True]
hot_tubs_df.head()
```

```python
# get the average review of each category
avg_accuracy = hot_tubs_df['review_scores_accuracy'].mean()
avg_location = hot_tubs_df['review_scores_location'].mean()
avg_cleanliness = hot_tubs_df['review_scores_cleanliness'].mean()
avg_checkin = hot_tubs_df['review_scores_checkin'].mean()
avg_communication = hot_tubs_df['review_scores_communication'].mean()
avg_value = hot_tubs_df['review_scores_value'].mean()

print(np.round(avg_accuracy, decimals=1))
print(np.round(avg_location, decimals=1))
print(np.round(avg_cleanliness, decimals=1))
print(np.round(avg_checkin, decimals=1))
print(np.round(avg_communication, decimals=1))
print(np.round(avg_value, decimals=1))

y_axis_ratings = [avg_accuracy, avg_location, avg_cleanliness, avg_checkin, avg_communication, avg_value]
x_axis = [1, 2, 3, 4, 5, 6]
tick_locations = [value for value in x_axis]
plt.figure(figsize=(10,5))
plt.xticks(tick_locations, ["Accuracy", "Location", "Cleanliness", "Check In", "Communication", "Value"])

plt.bar(x_axis, y_axis_ratings, color='r', alpha=0.5, align="center")

```

```python
# get the average review of each category to 3 decimal places to zoom into the 4-star ratings (ENHANCE!)
avg_accuracy = hot_tubs_df['review_scores_accuracy'].mean()
avg_location = hot_tubs_df['review_scores_location'].mean()
avg_cleanliness = hot_tubs_df['review_scores_cleanliness'].mean()
avg_checkin = hot_tubs_df['review_scores_checkin'].mean()
avg_communication = hot_tubs_df['review_scores_communication'].mean()
avg_value = hot_tubs_df['review_scores_value'].mean()

print(np.round(avg_accuracy, decimals=3))
print(np.round(avg_location, decimals=3))
print(np.round(avg_cleanliness, decimals=3))
print(np.round(avg_checkin, decimals=3))
print(np.round(avg_communication, decimals=3))
print(np.round(avg_value, decimals=3))

y_axis_ratings = [avg_accuracy, avg_location, avg_cleanliness, avg_checkin, avg_communication, avg_value]
x_axis = [1, 2, 3, 4, 5, 6]
tick_locations = [value for value in x_axis]
plt.figure(figsize=(10,5))
plt.xticks(tick_locations, ["Accuracy", "Location", "Cleanliness", "Check In", "Communication", "Value"])
plt.ylim(4.0, 5.0)

plt.bar(x_axis, y_axis_ratings, color='r', alpha=0.5, align="center")
```

```python
# get the average review of each category to 5 decimal places to zoom into the 4-star ratings (NEEDS MOAR ENHANCE!)

avg_accuracy = hot_tubs_df['review_scores_accuracy'].mean()
avg_location = hot_tubs_df['review_scores_location'].mean()
avg_cleanliness = hot_tubs_df['review_scores_cleanliness'].mean()
avg_checkin = hot_tubs_df['review_scores_checkin'].mean()
avg_communication = hot_tubs_df['review_scores_communication'].mean()
avg_value = hot_tubs_df['review_scores_value'].mean()

print(np.round(avg_accuracy, decimals=5))
print(np.round(avg_location, decimals=5))
print(np.round(avg_cleanliness, decimals=5))
print(np.round(avg_checkin, decimals=5))
print(np.round(avg_communication, decimals=5))
print(np.round(avg_value, decimals=5))

y_axis_ratings = [avg_accuracy, avg_location, avg_cleanliness, avg_checkin, avg_communication, avg_value]
x_axis = [1, 2, 3, 4, 5, 6]
tick_locations = [value for value in x_axis]
plt.figure(figsize=(10,5))
plt.xticks(tick_locations, ["Accuracy", "Location", "Cleanliness", "Check In", "Communication", "Value"])
plt.ylim(4.73, 4.89)

plt.bar(x_axis, y_axis_ratings, color='r', alpha=0.5, align="center")
```

```python
# get the count of 1, 2, 3, 4, and 5 star ratings for ACCURACY
num_accuracy = hot_tubs_df['review_scores_accuracy'].count()

one_star_accuracy_df = hot_tubs_df[(hot_tubs_df['review_scores_accuracy'] >= 0) & (hot_tubs_df['review_scores_accuracy'] < 1)].count()
num_1star_accuracy = one_star_accuracy_df['review_scores_accuracy']

two_star_accuracy_df = hot_tubs_df[(hot_tubs_df['review_scores_accuracy'] >= 1) & (hot_tubs_df['review_scores_accuracy'] < 2)].count()
num_2star_accuracy = two_star_accuracy_df['review_scores_accuracy']

three_star_accuracy_df = hot_tubs_df[(hot_tubs_df['review_scores_accuracy'] >= 2) & (hot_tubs_df['review_scores_accuracy'] < 3)].count()
num_3star_accuracy = three_star_accuracy_df['review_scores_accuracy']

four_star_accuracy_df = hot_tubs_df[(hot_tubs_df['review_scores_accuracy'] >= 3) & (hot_tubs_df['review_scores_accuracy'] < 4)].count()
num_4star_accuracy = four_star_accuracy_df['review_scores_accuracy']

five_star_accuracy_df = hot_tubs_df[(hot_tubs_df['review_scores_accuracy'] >= 4) & (hot_tubs_df['review_scores_accuracy'] <= 5)].count()
num_5star_accuracy = five_star_accuracy_df['review_scores_accuracy']


print(num_1star_accuracy)
print(num_2star_accuracy)
print(num_3star_accuracy)
print(num_4star_accuracy)
print(num_5star_accuracy)
```

```python
# reduce data to 5-star reviews for ACCURACY
five_star_accuracy_df = hot_tubs_df.loc[(hot_tubs_df['review_scores_accuracy'] == 5)].reset_index()
num_five_star_accuracy_locations = five_star_accuracy_df['id'].count()

# full map of 5-star reviews for ACCURACY
five_star_accuracy_map = five_star_accuracy_df.hvplot.points(
    "longitude",
    "latitude",
    geo = True,
    tiles = "OSM",
    frame_width = 700,
    frame_height = 500,
    scale = 0.5,
    color = "neighbourhood_cleansed", 
    title = f"5-Stars for Accuracy: {num_five_star_accuracy_locations} Locations"
)

# Display the map plot
five_star_accuracy_map
```
